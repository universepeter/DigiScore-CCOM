# 1. transform stream to .edf file

```python
import numpy as np
from pyedflib import highlevel
```

```python
signals= np. random.rand(5, 256*300)*200 # 5 minutes of random signal 
# ndarray can be obtained from emotive stream 
```

```python
channel_names = ['ch1', 'ch2', 'ch3', 'ch4', 'ch5']
signal_headers = highlevel.make_signal_headers(channel_names, sample_frequency=256)
header = highlevel.make_header(patientname='randomman', gender='Female')
highlevel.write_edf('random.edf', signals, signal_headers, header)
```
